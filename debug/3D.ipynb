```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function which returns subset or r length from n
from itertools import combinations

comb = list(combinations(np.arange(start=0, stop=10, step=1), 3))
#print(comb)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

N1 = 10
N2 = 10
N3 = 10

ma = np.zeros([10,10,10])
#ma[0,0,0] = 1

for c in comb:
    n1 = c[0]
    n2 = c[1]
    n3 = c[2]
    ma[n1,n2,n3] = 1
    #print(n1,n2,n3)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_aspect('auto')

ax.voxels(ma, edgecolor="k")

plt.show()
```
