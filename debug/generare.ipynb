```python
# Function which returns subset or r length from n
from itertools import combinations
import numpy as np
import math
```

```python
m = 11
arr = np.arange(start=1, stop=m+1, step=1)
c6 = list(combinations(arr, 6))
```

```python
for c in c6:
    print(f"{c[0]},{c[1]},{c[2]},{c[3]},{c[4]},{c[5]}")
```
